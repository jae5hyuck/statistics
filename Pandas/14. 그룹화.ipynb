# 14. 그룹화
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.groupby('학교')
```

```python
df.groupby('학교').get_group('북산고')
```

```python
df.groupby('학교').get_group('능남고')
```

```python
df.groupby('학교').mean() 
```

```python
df.groupby('학교')[['키', '국어']].mean()
```

```python
# df.groupby('학교').mean() 계산 가능한 데이터들의 평균값을 보여줘야하는데 내 파이썬 버전에서는 안된다. 학교가 문자열이라고 안된데...
```

```python
df.groupby('학교').size() # 각 그룹의 크기
```

```python
df.groupby('학교').size()['능남고'] # 학교로 그룹화를 한 뒤에 능남고에 해당하는 데이터의 수
```

```python
df.groupby('학교')['키'].mean()
```

```python
df.groupby('학교')[['국어','영어','수학']].mean()
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
df.groupby(['학교','학년']).mean(numeric_only = True) # 학교별 , 학년별 평균 데이
```

```python
df.groupby('학년').mean(numeric_only = True) # 학년별 평균 데이터
```

```python
df.groupby('학년').mean(numeric_only = True).sort_values('키')
```

```python
df.groupby('학년').mean(numeric_only = True).sort_values('키', ascending = False)
```

```python
df.groupby(['학교','학년']).sum(numeric_only=True)
```

```python
df.groupby('학교')[['이름','SW특기']].count() # 학교로 그룹화를 한 뒤에 각 학교별 SW특기 데이터의 수를 가져옴
```

```python
school = df.groupby('학교')
school['학년'].value_counts() # 학교로 그룹화를 한 뒤에 학년별 학생 수를 가져옴
```

```python
school['학년'].value_counts().loc['북산고'] # 학교로 그룹화를 한 뒤에 북산고에 대해서 학년별 학생 수를 가져옴
```

```python
school['학년'].value_counts().loc['능남고']
```

```python
school['학년'].value_counts(normalize=True).loc['북산고'] # 학생들의 수 데이터를 퍼센트로 비교하여 가져옴
```

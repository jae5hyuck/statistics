# 7. 데이터 선택 (loc)
이름을 이용해서 원하는 row 에서 원하는 col 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.loc['1번'] # index 1번에 해당하는 전체 데이터
```

```python
df.loc['국어'] # 이렇게는 안됨 왜냐, 바로 열을 가져 올 수 없음.
```

```python
df.loc['5번']
```

```python
df.loc['1번','국어'] # index 1번에 해당하는 국어 데이터
```

```python
df.loc[['1번','2번'],'영어'] # index 1번, 2번에 해당하는 영어 데이터
```

```python
df.loc[['1번','2번'],['영어','수학']]
```

```python
df.loc['1번':'5번','국어':'사회'] # index 1번부터 5번까지 (5번 포함 데이터를 말함), 국어부터 사회까지 데이터 
```

```python
df
```

```python
df.iloc[0:5,3:7]
```

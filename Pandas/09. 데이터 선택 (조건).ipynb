# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 #학생들의 키가 185 이상인지 여부를 true/false
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # filter 를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']]
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교']=='북산고')]
```

```python
df.loc[(df['키'] >= 185) & (df['학교']=='북산고'),['이름','수학']]
```

### | 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200 )] 
```

### str 함수

```python
filt = df['이름'].str.startswith('송')  # 송씨 성을 가진 사람
df[filt]
```

```python
df.loc[filt,['이름','학교','키']]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 사람
df[filt]
```

```python
filter = df['이름'].str.contains('송')
df.loc[filter,['이름','학교','키']]
```

```python
df[~filt] # 이름에 '태'가 들어가는 사람을 제외
```

```python
langs = ['Python','Java'] # 리스트 데이터를 만들어논것
filt = df['SW특기'].isin(langs) # SW특기가 Python 이거나 Java 인 사람
df[filt]
```

```python
df
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt] # 오류가 난 이유는 밑에서 설명
```

```python
df['SW특기'].str.lower().isin(langs) # 얘는 bowl이 다 정확하게 나오는 반면
```

```python
df['SW특기'].str.contains('Java') # 이 구문은 NaN으로 나오는 것도 있어서 filt 했을 때 오류가 뜬다
# filt 는 true/false 관해 true 만 보여주기 때문이다.
```

```python
df['SW특기'].str.contains('Java', na = True) # NaN 데이터에 대해서 True 로 설정
```

```python
df['SW특기'].str.contains('Java', na = False)
```

```python
filt = df['SW특기'].str.contains('Java',na=False)
df[filt]
```

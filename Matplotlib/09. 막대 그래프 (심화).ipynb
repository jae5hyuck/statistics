# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib  # 한글 폰트와 마이너스 사용을 위해 기본적으로 해줘야 하는 것
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.barh(labels, values)
plt.xlim(180,195)
```

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ...
```

```python
bar = plt.bar(labels,values)
plt.ylim(175,195)
for idx,rect in enumerate(bar):
    plt.text(idx,rect.get_height() + 0.5,values[idx],ha='center',color='b')   # 아직 이해 안됐음.
```

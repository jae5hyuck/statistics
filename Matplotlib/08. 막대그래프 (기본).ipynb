# 8.막대그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib  # 한글 폰트와 마이너스 사용을 위해 기본적으로 해줘야 하는 것
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.bar(labels, values, color ='r')
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키
colors = ['r','g','b']
plt.bar(labels, values, color = colors,alpha=0.3)  # alpha : 투명도 0-1
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.bar(labels, values)
plt.ylim(175,195)  # y축 설정
```

```python
plt.bar(labels, values,width=0.25)
```

```python
plt.bar(labels, values,width=0.3)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45)
plt.show()
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키
ticks=['1번학생','2번학생','3번학생']

plt.bar(labels,values,color='r')
plt.xticks(labels, ticks, rotation=90)
```

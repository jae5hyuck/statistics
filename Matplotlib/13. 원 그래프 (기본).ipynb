# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib  # 한글 폰트와 마이너스 사용을 위해 기본적으로 해줘야 하는 것
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javaschript','C#', 'C/C++', 'ETC']

plt.pie(values,labels= labels, autopct='%.1f%%',counterclock=False,startangle= 90)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javaschript','C#', 'C/C++', 'ETC']
# explode = [0.2,0.1,0,0,0,0]
explode=[0.05] * 6

plt.pie(values,labels=labels,explode=explode)
plt.show()
```

```python
plt.pie(values, labels =labels ,explode =explode)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
plt.pie(values, labels =labels ,explode =explode)
plt.title('언어')
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
plt.show()
```
